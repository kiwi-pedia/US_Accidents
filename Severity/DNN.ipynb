```python
def scaling(df):
    df_norm = df.copy()
    for column in df_norm.columns:
        df_norm[column] = (0.6*(df_norm[column] - df_norm[column].min()) / (df_norm[column].max() - df_norm[column].min()))+0.2
    return df_norm
```

```python
def logScaling(df):
    df_norm = df.copy()
    for column in df_norm.columns:
        df_norm[column] = (df_norm[column]+1.0)
    return df_norm
```

# DNN Model

```python
from sklearn.neural_network import MLPClassifier
DNN = MLPClassifier(hidden_layer_sizes=(5,5,5,5,5,5),  #預設值1層,數目100
         activation='relu',  #{'identity'無，'logistic'為sigmoid，'tanh'，'relu'}，預設='relu'
         solver='adam',   #{'lbfgs'牛頓，'sgd'隨機梯度，'adam'隨機梯度優化}，預設='adam'
         batch_size='auto',  #'auto'為min(200～樣本數)
         learning_rate='adaptive',  #constant固定'，'invscaling隨著時間遞減'，'adaptive誤差減少時不改變'}，預設 ='constant' 
         learning_rate_init=0.001, #優化器為sgd、adam使用
         power_t=0.5,     # 優化器為sgd，則可以設定強化學習率   
         max_iter=1000,    #訓練次數
         shuffle=True,    #隨機設定權重
         random_state=1,   #隨機的基礎值
         momentum=0.9,
         # tol=1e-5
         # early_stopping=True          
                   )
```

```python
DNN.fit(X_train, y_train)
```

```python
DNN_pred = DNN.predict(X_test)
DNN_pred
```

```python
DNNtest_score=DNN.score(X_test,y_test)
print('The Accuracy of DNN Classifier on testing set:', DNNtest_score)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,DNN_pred))
```
